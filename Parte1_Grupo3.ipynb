```
import pandas as pd   
import matplotlib.pyplot as plt  
import seaborn as sns

df=pd.read_csv('student-mat.csv')
```

```
for colunas in df.columns:
    sns.countplot(colunas,data=df)
    plt.show()
```

```
circular = plt.subplots(figsize=(4, 4))
plt.pie(df['school'].value_counts().tolist(), 
        labels=['GP', 'MS'],
        autopct='%1.1f%%', startangle=0)
axis = plt.axis('equal')
```

```
circular = plt.subplots(figsize=(4, 4))
plt.pie(df['Pstatus'].value_counts().tolist(), 
        labels=['Juntos', 'Separados'], 
        autopct='%1.1f%%', startangle=0)
axis = plt.axis('equal')
```

```
circular = plt.subplots(figsize=(4, 4))
plt.pie(df['schoolsup'].value_counts().tolist(), 
        labels=['nao', 'sim'], 
        autopct='%1.1f%%', startangle=0)
axis = plt.axis('equal')
```

```
circular = plt.subplots(figsize=(4, 4))
plt.pie(df['higher'].value_counts().tolist(), 
        labels=['sim', 'nao'], 
        autopct='%1.1f%%', startangle=0)
axis = plt.axis('equal')
```
