```
import pandas as pd
import statsmodels.api as sm  

df=pd.read_csv('student-mat.csv')
df=df[df.G3 != 0]
```

```
df.columns
```

```
#School

df1=df[['school','G3' ]]
dummy=pd.get_dummies(df1['school'],drop_first=True)
df1=pd.concat([df1,dummy], axis=1)
y=df1['G3'] 
x=df1[['MS']]
                                  
X1=sm.add_constant(x)                                                
model = sm.OLS(y,X1).fit()          
model.summary()



```
